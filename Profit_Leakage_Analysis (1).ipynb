\# *Business Problem

The company offers discounts to increase sales, but excessive discounts may reduce profitability.

This project analyzes sales, profit, discounts, customers, and regional performance to identify:

loss-making products
risky discount strategies
profit leakage areas
high-risk transactions **bold text*


============================VISUALIZATIONS=================================

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar',color='green')

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()
```

**INSIGHT**

Technology category contributes the highest sales among all categories.

```python
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='bar',color='yellow')

plt.title("Profit by Category")
plt.ylabel("Profit")
plt.show()
```

INSIGHTS

Some categories generate strong sales but comparatively lower profits.

```python
sns.scatterplot(x='Discount', y='Profit', data=df,color='purple')

plt.title("Discount vs Profit")
plt.show()
```

INSIGHT

Higher discount values are frequently associated with negative profits, indicating discount leakage.

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(color='green')

plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.show()
```

INSIGHT

Sales fluctuate across different months, showing seasonal purchasing patterns.

```python
region_profit = df.groupby('Region')['Profit'].sum()

region_profit.plot(kind='bar',color='green')

plt.title("Region-wise Profit")
plt.ylabel("Profit")
plt.show()
```

INSIGHT

Some regions perform significantly better in profitability compared to others.

```python
df['Risk Score'] = 0

df.loc[df['Discount'] > 0.4, 'Risk Score'] += 3

df.loc[df['Profit'] < 0, 'Risk Score'] += 5

#RISK SCORE
```

```python
def risk_level(score):
    if score >= 5:
        return "High Risk"
    elif score >= 3:
        return "Medium Risk"
    else:
        return "Low Risk"

df['Risk Level'] = df['Risk Score'].apply(risk_level)

#RISK LEVEL
```

```python
sns.countplot(
    x='Category',
    hue='Risk Level',
    data=df,
    palette=['green', 'red']
)

plt.title("Risk Level by Category")
plt.figure(figsize=(8,5))

plt.show()
```

INSIGHT

Furniture category contains a relatively higher proportion of high-risk transactions compared to Technology, suggesting that discounting and low-profit margins may be affecting profitability in this segment.

```python
sns.heatmap(
    df[['Sales','Profit','Discount','Quantity']].corr(),
    annot=True
)

plt.title("Correlation Heatmap")
plt.show()
```

INSIGHT

Discount shows negative correlation with profit, confirming that aggressive discounting impacts profitability.

```python
loss_products = df.groupby('Product Name')['Profit'].sum() #total profit for each product
```

```python
loss_products = loss_products.sort_values().head(10) # top 10 products with biggest losses
```

```python
plt.figure(figsize=(10,6))

loss_products.plot(kind='barh', color='red')

plt.title("Top 10 Loss-Making Products")

plt.xlabel("Profit")

plt.ylabel("Product Name")
plt.tight_layout()

plt.show()
```

INSIGHT

Several products generated repeated losses despite strong sales activity, indicating potential pricing inefficiencies or excessive discounting.
